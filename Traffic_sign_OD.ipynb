```python
from detecto import core, utils, visualize
```

```python
dataset = core.Dataset('Traffic-small/')
```

```python
print(dataset)
```

```python
image, target = dataset[1]
print(image, target)
```

```python
model = core.Model(['SL', 'MSL'])

model.fit(dataset, epochs=30, lr_step_size=5, learning_rate=0.001)
```

```python
model.save('TS_weights.pth')
```

```python
# Specify the path to your image
image = utils.read_image('Traffic-small/00000_00001.jpg')
predictions = model.predict_top(image)

# predictions format: (labels, boxes, scores)
labels, boxes, scores = predictions

print(labels)
print(boxes)
print(scores)
```

```python
image_pred = visualize.show_labeled_image(image, boxes, labels)
```

```python
import glob
import os
import cv2  
import matplotlib.pyplot as plt
import pandas as pd
base_dir='Traffic-small/'
rows=[]
for file in glob.glob(os.path.join(base_dir, '*.png')):
    
    row ={}
    filname = os.path.basename(file)
    filename = os.path.splitext(filname)[0]
    image = utils.read_image(file)
    predictions = model.predict_top(image)
    labels, boxes, scores = predictions
    x,y,w,h = boxes[0]
    label = labels[0]
    score = scores[0]
    row['filename']=filname
    row['x'] = float(x)
    row['y'] = float(y)
    row['w'] = float(w)
    row['h'] = float(h)
    row['pred_class'] = label
    row['score'] = float(score)
    
    cv2.rectangle(image, (x, y), (w, h), (255,255,255), 1)
    cv2.putText(image, '%s: %.3f' % (label, score), (x+1, y+6), cv2.FONT_HERSHEY_SIMPLEX,
                        0.25, (255, 255, 255), thickness=1)
    cv2.imwrite('TFresult/'+filename+'bbox.jpg', image)
    rows.append(row)
    df = pd.DataFrame(rows)
    df = df.set_index('filename')
    df.to_csv('TF_pred.csv')
    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    ax.imshow(image)
```
